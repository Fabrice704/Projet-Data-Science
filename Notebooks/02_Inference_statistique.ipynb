##  correlation de pearson

```python
variable_quanti = df.select_dtypes(include = "number").columns.tolist()
variable_quanti
```

```python
df[variable_quanti]
```

```python
corr_matrix = df[variable_quanti].corr()
corr_matrix
```

```python
from scipy import stats
```

```python
# creere mon dataframe vide pour stoquer les p_values
p_values_matrix = pd.DataFrame(index = variable_quanti, columns = variable_quanti)

# calculer la correlation et la p_value pour cahque paire de variable
# avec la fonction .loc j'accede a une cellule specitique a parti du nom des ligne et de colonne dans le dataframe p_values_matrix que je viens de creer
# cellule = correspond a l'entersection de var1 et var2 que je rempli avec la p_value correspondant

for var1 in variable_quanti:
    for var2 in variable_quanti:
        if var1 != var2:
            statistic, p_value = stats.pearsonr(df[var1], df[var2])
            p_values_matrix.loc[var1, var2] = p_value
        else:
            # la correlation d'une variable avec elle meme est toujours 1 donc pas besoin de la pevaleur
            p_values_matrix.loc[var1, var2] = 1

# afficher ma matrice de p_value
p_values_matrix
            
```

## test de normalité

```python
# de notre analyque numerique et graphique univarié on a vu que nos variables semblent ne pas etre normalement distribuées
# verifians cela via le test

# test de shapiro-wilk

shapiro_results = {}
for i in variable_quanti:
    shapiro_test = stats.shapiro(df[i])
    shapiro_results[i] = shapiro_test

for key, value in shapiro_results.items():
    print (f"Variable : {key},    Test de shapiro : {value}")

    
```

COCLUSION:

avec les p_value de mes variables très inferieur a 0,05, l'ypothese d'ANORMALITE de mes variables est STATISTIQUMENT SIGNIFICATIVE
donc aucune de mes variables ne suit une distribution normale

notre modalité a 2 variables , on a donc le test de t_student

## entre ma variable cible (MEDV) et ma variable quali(CHAS)

## test de t_student(test parametrique )

### c'est un test parametrique qui repose sur la normalité et l'egalité des variance

le test de t student compare les moyennes de 2 groupes il suppose que les variances sont egales

```python
valeur_maison_rivière = df[df.CHAS == 1]["MEDV"]
valeur_maison_pas_rivière = df[df.CHAS == 0]["MEDV"]
stats.ttest_ind(valeur_maison_rivière, valeur_maison_pas_rivière)
```

conclusion: p_value < 0,05 donc la H0 rejeté la moyenne dans les deux groupes est differente

effectivement sur le graphique selon notre analyse bivarié on voit que les maison qui sont dans un quartier qui borde la rivière ont majoritairement un valeur superieur para rapport au maisons dans un quartier qui ne borde pas la riviere
et grace alla p_value cette difference est STATISTIQUEMENT SIGNIFICATIVE

Donc la proximité a la rivière (CHAS) est une variable qui impact la valeur de la maison ( variable cible)

```python
# ce test repose sur la normalité et l'egalité des variances / verifions la normalité des sous-groupes
```

```python
stats.shapiro(valeur_maison_rivière)
```

p_value < 0,05 donc l'ypothese de H0 est rejeté, les données ne suivent pas une distribution Normale

```python
stats.shapiro(valeur_maison_pas_rivière)
```

p_value < 0,05 donc l'ypothese de H0 est rejeté, les donnees ne suivent pas une distribution Normale

```python
# verifions aussi l'egalité des variances entre les 2 groupes
# on a le choix entre  test de levene ou le test de fligner-killer ( tous les 2 sont des test non parametrique qui ne reposent pas sur la normalité) 
#donc trèes adapté dans notre cas

stats.levene(valeur_maison_rivière, valeur_maison_pas_rivière)

# a cote de sa on a le test de barlett( qui par contre repose sur la normalité)
```

p_value < 0.05 donc l'ypothese nulle H0 = (les variances dans les deux groupes sont identique) est rejeté 


les variances des sous groupes sont differente/ donc aucune conditions n'est verifié ni la normalité, ni l'egalité des variances
donc le test de t-student n'est pas  adapté/ je ne peu me fier a cela

pour cela j'opte pour un test NON PARAMETRIQUE qui ne repose pas sur la normalité


## test de wilcoxon ( test non_ parametrique)

```python
stats.mannwhitneyu(valeur_maison_rivière, valeur_maison_pas_rivière )
```

conclusion: p_value < 0,05 l'ypothese H0 rejetée , les moyennes dans les 2 groupes sont differentes
